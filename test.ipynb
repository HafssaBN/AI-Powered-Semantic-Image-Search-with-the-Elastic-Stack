```python
import os
import re
import requests
from bs4 import BeautifulSoup

# List of URLs from your SerpApi results
urls = [
    "https://www.airbnb.com/rooms/53457746",
    "https://www.airbnb.com/rooms/52067417",
    "https://www.agoda.com/majorelle-hostel/hotel/marrakech-ma.html",
    "https://www.booking.com/hotel/ma/majorelle-hostel.html",
    "https://www.hotelsmarrakesh.org/en/property/majorelle-hostel/rooms.html",
    "https://www.travelmyth.com/Marrakech/Hotels/Riad-Green-Vines_tmid1652568",
    "https://majorellehostel.morocco-ma.website/en/",
    "https://en.planetofhotels.com/morocco/marrakech/majorelle-hostel",
    "https://www.instagram.com/riadgreenvines"
]

# Folder to save images
output_dir = "riad_greenvines_images"
os.makedirs(output_dir, exist_ok=True)

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115 Safari/537.36"
}

def download_image(img_url, folder):
    try:
        img_data = requests.get(img_url, headers=headers, timeout=10).content
        filename = os.path.join(folder, img_url.split("/")[-1].split("?")[0])
        with open(filename, "wb") as f:
            f.write(img_data)
        print(f"‚úÖ Downloaded: {filename}")
    except Exception as e:
        print(f"‚ùå Failed to download {img_url} ‚Äî {e}")

for url in urls:
    try:
        print(f"\nüåç Scraping {url}")
        resp = requests.get(url, headers=headers, timeout=15)
        soup = BeautifulSoup(resp.text, "html.parser")

        # Extract all image links
        imgs = set()
        for img in soup.find_all("img", src=True):
            src = img["src"]
            if not src.startswith("http"):
                src = requests.compat.urljoin(url, src)
            if re.search(r"\.(jpg|jpeg|png|webp)", src, re.IGNORECASE):
                imgs.add(src)

        print(f"Found {len(imgs)} images on {url}")

        # Download all images
        for img_url in imgs:
            download_image(img_url, output_dir)

    except Exception as e:
        print(f"‚ö†Ô∏è Error scraping {url}: {e}")
```
